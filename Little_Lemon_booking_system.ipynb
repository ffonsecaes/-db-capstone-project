```python
#Connection to Little Lemon DB
try:
    connection = connector.connect(
        user = 'root',
        password = 'Contrasena',
        db = 'littlelemondb'
    )
    print('Connection Successfully')
except Exception as e:
    print('Please review your connectivity parameters')
    print(e)
```

```python
#Query all tables in the database
try:
    cursor = connection.cursor()
    tables = cursor.execute('SHOW TABLES')

    for table in cursor:
        print(table)
except Exception as e:
    print(e)
    
```

```python
#JOIN two tables in the database
try:
    cursor.execute("""
    SELECT orders.total_cost, customers.name, customers.phone, customers.email 
    FROM orders 
    INNER JOIN customers 
    ON orders.customer_id = customers.customer_id 
    WHERE orders.total_cost > 60 
    ORDER BY orders.total_cost DESC;
    """)

    print(cursor.column_names)
    
    customers = cursor.fetchall()
    for customer in customers:
        print(customer)
        
except Exception as e:
    print(e)
```
